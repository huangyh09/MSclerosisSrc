```python
import matplotlib
import matplotlib.pyplot as plt

from pyseqlib.plot import plot_gene, set_frame

import hilearn
hilearn.plot.set_style(grid_alpha=False)

import seaborn
color4 = seaborn.color_palette("hls", 4)
color8 = seaborn.color_palette("hls", 8)

fig = plt.figure(figsize=(10, 1), dpi=300)
ax = plt.subplot(1,1,1)
exon_emph = [[139109024, 139109719],
             [139060338, 139061138],
            
             [139109024, 139109648],
             [139043520, 139047853],
             [139053451, 139053581],
             [139053965, 139054095],
             [139055205, 139055295],
             [139061036, 139061138]
            ]

# exon_colors = [color4[0]] * 2 + [color4[1]] * 6

exon_colors = ['#df5858'] * 2 + ['#488a99'] * 6

plot_gene(genes[0], exon_emph=exon_emph, exon_colors=exon_colors)

plt.scatter([139047152], [-1.2], s=40, c='grey')

# plt.arrow(139047152, 2, 0, -1, width=10)

plt.title("ZC3HAV1")
set_frame(ax)

# fig.savefig("ZC3HAV1_gene.png",dpi=300, bbox_inches='tight')
# fig.savefig("ZC3HAV1_gene.pdf",dpi=300, bbox_inches='tight')
plt.show()
```
