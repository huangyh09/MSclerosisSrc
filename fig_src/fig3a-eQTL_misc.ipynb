## Summary

```python
df_stats = pd.read_csv(dat_dir + 'eQTL_summary.txt', sep='\t')
idx = np.argsort(df_stats['# eQTL'].values)[::-1]
df_stats = df_stats.loc[idx]
df_stats.index = range(df_stats.shape[0])
```

```python
df_stats['Overall rate'][df_stats['Overall rate'] == '-'] = '0%'
df_stats['Overall rate'] = [float(x[:-1]) / 100 for x in 
                            df_stats['Overall rate']]

df_stats['Overall rep'] = np.round(
    df_stats['Overlap Overall'] * df_stats['Overall rate'])
```

```python
df_stats
```

```python
import matplotlib.pyplot as plt

novel_cells = list(df_stats['Cell type'][df_stats['Novel']])

x = np.arange(df_stats.shape[0])  # the label locations
width = 0.35  # the width of the bars

fig = plt.figure(figsize=(6, 4),dpi=300)
rects1 = plt.bar(x - width/2, df_stats['# eQTL'], width, label='# Our hits')
rects3 = plt.bar(x + width/2, df_stats['Overlap Overall'], width, 
                 color='gainsboro', label='# Unreplicated')
rects2 = plt.bar(x + width/2, df_stats['Overall rep'], 
                 width, label='# Replicated')


# rects3 = plt.bar(x + width/2, df_stats['Overlapped'], width, 
#                  color='gainsboro', label='# Unreplicated')
# rects2 = plt.bar(x + width/2, df_stats['Replicated Oelen'], 
#                  width, label='# Replicated')

for i in range(len(x)):
    _color = 'tab:orange'
    if df_stats['Cell type'][i] in novel_cells:
        _text = 'Novel'
        _color = 'red'
    elif df_stats['Overall rate'][i] > 0:
        _text = '%d%%' %(df_stats['Overall rate'][i] * 100)
    else:
        _text = 'NA'
    plt.text(x[i], df_stats['# eQTL'][i] + 10, _text, c=_color, ha='center')

plt.ylim(0, 490)
plt.xticks(x, df_stats['Cell type'], rotation=30, ha='right')
plt.ylabel('#eQTLs')
plt.legend()
plt.tight_layout()
plt.grid(alpha=0.3)

plt.tight_layout()
# plt.savefig(fig_dir + 'global_eQTL_summary.pdf')
plt.show()
```

## Colocalization plot

```python
df_gwas = pd.read_csv(dat_dir + 'ColocPlotData/SelectedGWASSumStats.txt', 
                      sep='\t')
df_eqtl = pd.read_csv(dat_dir + 'ColocPlotData/SelectedQTLSumStats.CD8+_T_cell.txt', 
                      sep='\t')
```

```python
df_gwas.head()
```

```python
df_eqtl.head()
```

```python
_col_gene = 'ZC3HAV1'

_idx2 = df_eqtl.feature_id == _col_gene
_post = np.median(df_eqtl.snp_position[_idx2].values)
_chrom = df_eqtl.feature_chromosome[_idx2].values[0]

_idx1 = ((df_gwas.CHR == _chrom) * 
         (df_gwas.BP >= _post - 100000) *
         (df_gwas.BP <= _post + 100000)
        )
```

```python
sum(_idx1)
```

```python
fig = plt.figure(figsize=(5, 4.5), dpi=300)
plt.subplot(2, 1, 1)
plt.scatter(df_gwas.BP[_idx1]/1000, 
            -np.log10(df_gwas.P[_idx1]), 
            color='lightseagreen', s=5)
plt.xlim(138729-110, 138729+110)
plt.xticks([])
plt.ylabel('-log10(P), GWAS')

plt.subplot(2, 1, 2)
plt.scatter(df_eqtl.snp_position[_idx2] / 1000, 
            -np.log10(df_eqtl.p_value[_idx2]), 
            color='skyblue', s=5)
plt.xlim(138729-110, 138729+110)
plt.xticks([138650, 138700, 138750, 138800])
plt.xlabel('Chromosome 7 (kb)')
plt.ylabel('-log10(P), eQTL')
plt.tight_layout()

plt.savefig(fig_dir + 'coloc_%s.pdf' %(_col_gene))
plt.show()
```
